```
import pandas as pd
import seaborn as sns

df1=pd.read_csv("data_1.csv", header=0)

print(len(df1))

print(df1.head())
```

```
ax=sns.countplot(x='EMOTIONAL HEALTH', data=df1)
total = float(len(df1))
ax.set_ylabel('count')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}'.format(height/total),
            ha="center") 
```
