```
import pickle
```

```
file1 = open('CNNBILSTM_Glove', 'rb')
history1 = pickle.load(file1)
file1.close()

file1 = open('BILSTM_glove', 'rb')
history2 = pickle.load(file1)
file1.close()

file1 = open('LSTMCNN_glove', 'rb')
history3 = pickle.load(file1)
file1.close()

file1 = open('LSTM_glove', 'rb')
history4 = pickle.load(file1)
file1.close()
```

```
import pickle
import matplotlib as mpl
%matplotlib inline
from matplotlib import pyplot as plt
from keras.utils import plot_model

fig = plt.figure(figsize=(10,9))
ax = fig.add_subplot(111)
ax.plot(history1['val_accuracy'])
ax.plot(history2['val_accuracy'])
ax.plot(history3['val_accuracy'])
ax.plot(history4['val_accuracy'])
plt.ylabel('Accuracy', fontsize=18)
plt.xlabel('Epoch', fontsize=18)
plt.legend(['CNN+BILSTM', 'BILSTM','CNN+LSTM','LSTM'], loc='upper left',fontsize=13)

plt.savefig('glove_acc_1.png')
```
